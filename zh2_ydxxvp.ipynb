```sql
SELECT 
    s.szoba_id,
    YEAR(f.mettol) AS ev,
    COUNT(f.foglalas_pk) AS foglalas_szama,
    CASE 
        WHEN GROUPING(s.szoba_id) = 1 THEN 'Szoba összesen'
        ELSE 'Részösszeg'
    END AS osszeg_tipus
FROM Szoba s
LEFT JOIN Foglalas f ON s.szoba_id = f.szoba_fk
GROUP BY s.szoba_id, YEAR(f.mettol)
HAVING s.szoba_id IS NOT NULL OR GROUPING(s.szoba_id) = 1
ORDER BY s.szoba_id, YEAR(f.mettol);
```

```sql
SELECT 
    s.szoba_id,
    fb.foglalas_pk,
    DATEDIFF(day, fb.mettol, fb.meddig) AS foglalas_idotartam,
    CASE 
        WHEN pb.elozo_meddig IS NOT NULL THEN DATEDIFF(day, pb.elozo_meddig, fb.mettol)
        ELSE 0
    END AS elozo_idotartam
FROM Szoba s
LEFT JOIN Foglalas fb ON s.szoba_id = fb.szoba_fk
LEFT JOIN (
    SELECT 
        f1.szoba_fk,
        f1.foglalas_pk,
        f2.meddig AS elozo_meddig
    FROM Foglalas f1
    LEFT JOIN Foglalas f2 ON f1.szoba_fk = f2.szoba_fk AND f1.mettol > f2.mettol
) pb ON fb.szoba_fk = pb.szoba_fk AND fb.foglalas_pk = pb.foglalas_pk;
```
